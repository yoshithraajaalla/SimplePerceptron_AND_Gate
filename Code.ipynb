```python
# Importing Packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Initializing variables for AND Gate
X = np.asarray([[0,0],[0,1],[1,0],[1,1]])
W = np.asarray([1,1])
Wsum_Arr = np.dot(X,W)
Theta = 1.5

y_pred = np.asarray([0,0,0,0])
```

```python
# STEP Function
def step(NW,n):
  if (NW[n]>=Theta):
    return 1
  else:
    return 0
```

```python
for i in range(X.shape[0]):
  y_pred[i] = step(Wsum_Arr,i)

print(y_pred) #Expected Output: [0,0,0,1]
```

```python
slope = -1
intercept = Theta

x_plane = np.linspace(-2,2,100)
y_plane = slope * x_plane + intercept

plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.plot(x_plane, y_plane, '-')
plt.xlim([-1,2])
plt.ylim([-1,2])
plt.show()
```
